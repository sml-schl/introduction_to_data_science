## 1. Basic Python Operations

## 2. Working with Lists and NumPy Arrays

NumPy is a fundamental package for scientific computing in Python.

```python
import numpy as np

# Lists
numbers = [1, 2, 3, 4, 5]
print("Python List operations:")
print(f"Original list: {numbers}")
print(f"First element: {numbers[0]}")
print(f"List slicing: {numbers[1:3]}")
print(f"List length: {len(numbers)}")

# NumPy arrays
array = np.array(numbers)
print("\nNumPy array operations:")
print(f"Array: {array}")
print(f"Mean: {array.mean()}")
print(f"Sum: {array.sum()}")
print(f"Square root: {np.sqrt(array)}")
```

## 3. Data Visualization

Matplotlib is a plotting library for Python. Here's a simple visualization example:

```python
import matplotlib.pyplot as plt

# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# First subplot: Line plot
ax1.plot(x, y, 'b-', label='sin(x)')
ax1.set_title('Simple Line Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.legend()

# Second subplot: Scatter plot
ax2.scatter(numbers, np.square(numbers), color='red', label='x')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('x')
ax2.legend()

plt.tight_layout()
plt.show()
```
